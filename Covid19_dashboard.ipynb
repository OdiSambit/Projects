## Top affected countries analysis

### Subtask:
Identify and visualize the top N countries based on metrics like total cases, deaths, and recoveries.


**Reasoning**:
Sort the dataframe by 'Confirmed', 'Deaths', and 'Recovered' columns, select the top 10 rows for each and store them in new dataframes. Then create bar plots for each of the dataframes.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort by confirmed cases and select top 10
top_10_confirmed = df.sort_values(by='Confirmed', ascending=False).head(10)

# Create bar plot for top 10 confirmed cases
plt.figure(figsize=(12, 6))
sns.barplot(x='Country/Region', y='Confirmed', data=top_10_confirmed, palette='viridis')
plt.title('Top 10 Countries by Confirmed Cases')
plt.xlabel('Country/Region')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Sort by deaths and select top 10
top_10_deaths = df.sort_values(by='Deaths', ascending=False).head(10)

# Create bar plot for top 10 deaths
plt.figure(figsize=(12, 6))
sns.barplot(x='Country/Region', y='Deaths', data=top_10_deaths, palette='plasma')
plt.title('Top 10 Countries by Deaths')
plt.xlabel('Country/Region')
plt.ylabel('Deaths')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Sort by recovered cases and select top 10
top_10_recovered = df.sort_values(by='Recovered', ascending=False).head(10)

# Create bar plot for top 10 recovered cases
plt.figure(figsize=(12, 6))
sns.barplot(x='Country/Region', y='Recovered', data=top_10_recovered, palette='magma')
plt.title('Top 10 Countries by Recoveries')
plt.xlabel('Country/Region')
plt.ylabel('Recovered Cases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Death and recovery analysis

### Subtask:
Analyze the relationship between deaths and recoveries, perhaps looking at death rates or recovery rates.


**Reasoning**:
Calculate the death and recovery rates and add them as new columns to the dataframe, then create a scatter plot to visualize the relationship between these rates, using confirmed cases for size, add labels and a title, and display the plot.



```python
# Calculate death rate and recovery rate
df['Death Rate'] = (df['Deaths'] / df['Confirmed']) * 100
df['Recovery Rate'] = (df['Recovered'] / df['Confirmed']) * 100

# Create a scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Recovery Rate', y='Death Rate', size='Confirmed', hue='WHO Region', data=df, alpha=0.6, sizes=(20, 1000))

# Add labels and title
plt.title('Death Rate vs. Recovery Rate per Country')
plt.xlabel('Recovery Rate (%)')
plt.ylabel('Death Rate (%)')
plt.grid(True)
plt.legend(title='WHO Region')

# Display the plot
plt.show()
```

## Heatmap visualization

### Subtask:
Create heatmaps to visualize correlations between numerical features or to show the distribution of a specific metric across countries or over time.


**Reasoning**:
Select numerical columns, calculate the correlation matrix, and create a heatmap with annotations.



```python
# Select numerical columns for correlation analysis
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate the correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Add a title to the heatmap
plt.title('Correlation Heatmap of Numerical Features')

# Display the heatmap
plt.show()
```

## Combine visualizations for dashboard
