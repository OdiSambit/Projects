```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset("iris")
```

```python
df
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
X= df.drop('species',axis=1)
Y=df['species']
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors = 5)
model_knn.fit(X_train,Y_train)
```

```python
model_knn.score(X_test,Y_test)#1.0 means 100% which is overefit
```

```python
#svc
from sklearn.svm import SVC
model_svm=SVC(gamma='auto')# c is penalty
model_svm.fit(X_train,Y_train)
```

```python
model_svm.score(X_test,Y_test)
```

Now we will use Grid Search CV for a better score

```python
from sklearn.model_selection import GridSearchCV
classifier = GridSearchCV((model_svm),{
    'C':[1,10,20,30],
    'kernel':['rbf','linear']
},cv=5,return_train_score=False)
classifier.fit(X_train,Y_train)

```

```python
classifier.cv_results_
```

```python
results = pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
#so here you can see if you choose param_c as 1 kernel as rbf then score will be 95% this is best hyperparameter
```

```python
#Randomized search
from sklearn.model_selection import RandomizedSearchCV
classifier_r = RandomizedSearchCV((model_svm),{
    'C':[1,10,20,30],
    'kernel':['rbf','linear']
},cv=5,return_train_score=False,n_iter=4)
#n_iter - no of iterations
#cv - cross validation
```

```python
classifier_r.fit(X_train,Y_train)
```
